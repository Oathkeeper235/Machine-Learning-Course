# Loading the dataset



```python
import pandas as pd
import numpy as np

df = pd.read_excel("Data_Cortex_Nuclear.xls")
df
```

```python
df.dropna(inplace=True)
df
```

```python
x = df.drop(["MouseID", "Genotype", "Treatment", "Behavior", "class"], axis=1)
```

# Getting to know the dataset and preprocessing

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

x = StandardScaler().fit_transform(x)

pca = PCA(2)

x_pca = pca.fit_transform(x)
```

```python
import seaborn as sns

sns.relplot(x = x_pca[:, 0], y = x_pca[:, 1], hue= df["class"], palette="muted")
```

# Clustering

```python
import matplotlib.pyplot as plt
from sklearn.metrics import homogeneity_score, completeness_score

def visualize_clustered_data_2d(data_2d, labels):
    plt.figure(figsize = (10, 10))
    sns.color_palette("rocket")
    sns.scatterplot(x=data_2d[:,0], y=data_2d[:,1], hue=labels.astype(str))
```

```python
from sklearn.manifold import TSNE

homogeneity = {}
completeness = {}

tsne = TSNE(2)
x_tsne = tsne.fit_transform(x)

visualize_clustered_data_2d(x_tsne, df["class"])
```

## Gaussian Mixture Models

```python
from sklearn.mixture import GaussianMixture

gaussian_mixture = GaussianMixture(len(np.unique(df["class"])))
y_pred_gaussian = gaussian_mixture.fit_predict(x)

visualize_clustered_data_2d(x_tsne, y_pred_gaussian)
homogeneity["Gaussian Mixture"] = homogeneity_score(df["class"], y_pred_gaussian)
completeness["Gaussian Mixture"] = completeness_score(df["class"], y_pred_gaussian)
```

## K-Means

```python
from sklearn.cluster import KMeans

kmeans = KMeans(len(np.unique(df["class"])), n_init='auto')
y_pred_kmeans = kmeans.fit_predict(x)

visualize_clustered_data_2d(x_tsne, y_pred_kmeans)
homogeneity["KMeans"] = homogeneity_score(df["class"], y_pred_kmeans)
completeness["KMeans"] = completeness_score(df["class"], y_pred_kmeans)
```

## Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering

hierarchical = AgglomerativeClustering(len(np.unique(df["class"])))
y_pred_hierarchical = hierarchical.fit_predict(x)

visualize_clustered_data_2d(x_tsne, y_pred_hierarchical)
homogeneity["Hierarchical Clustering"] = homogeneity_score(df["class"], y_pred_hierarchical)
completeness["Hierarchical Clustering"] = completeness_score(df["class"], y_pred_hierarchical)
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=5, min_samples=5)
y_pred_dbscan = dbscan.fit_predict(x)

visualize_clustered_data_2d(x_tsne, y_pred_dbscan)
homogeneity["DBSCAN"] = homogeneity_score(df["class"], y_pred_dbscan)
completeness["DBSCAN"] = completeness_score(df["class"], y_pred_dbscan)
```

# Conclusion


## Homogeneity Score:

```python
sns.barplot(
    data=df, x=list(homogeneity.values()), y=list(homogeneity.keys()), 
    capsize=.4, errcolor=".5", linewidth=3, orient='h',
    edgecolor=".5", facecolor=(0, 0, 0, 0),
)
```

## Completeness Score:

```python
sns.barplot(
    data=df, x=list(completeness.values()), y=list(completeness.keys()), 
    capsize=.4, errcolor=".5", linewidth=3, orient='h',
    edgecolor=".5", facecolor=(0, 0, 0, 0),
)
```

### DBSCAN with (eps=5, min_samples=5) provided the best clustering results.
